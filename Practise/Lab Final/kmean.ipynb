```python
#K-Means Clustering

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

df = pd.read_csv("Wholesale customers data.csv")
df.head()
```

```python
#statistics of data
df.describe()
```

```python
#standardising the data
from sklearn.preprocessing import StandardScaler

df = pd.read_csv("Wholesale customers data.csv")
df.head()

scaler = StandardScaler()
data_scaled = scaler.fit_transform(df)

pd.DataFrame(data_scaled)

#statistics of scaled data
pd.DataFrame(data_scaled).describe()
```

```python
#defining the kmeans function with initialization as k-means++
kmeans = KMeans(n_clusters=2, init='k-means++')

#fitting the k means algorithm on scaled data
kmeans.fit(data_scaled)
```

```python
kmeans.inertia_
```

```python
#Inertia (SSE) measures how well a dataset was clustered by K-Means. It is calculated by
#measuring the distance between each data point and its centroid, squaring this distance, and
#summing these squares across one cluster. However, this is a tradeoff because as K increases,
#inertia decreases.

#fitting multiple k-means algorithms and sorting the values in an empty list
SSE = []

for cluster in range(1,20):
    kmeans = KMeans(n_clusters=cluster, init='k-means++')
    kmeans.fit(data_scaled)
    SSE.append(kmeans.inertia_)
    
#converting the results into a dataframe and plotting them
frame = pd.DataFrame({'Cluster': range(1,20), 'SSE': SSE})
plt.figure(figsize=(12,6))
plt.plot(frame['Cluster'], frame['SSE'], marker='o')
plt.xlabel("number of clusters")
plt.ylabel('inertia')
```

```python
#k-means using 5 clusters and k-mean++ initialization
kmeans = KMeans(n_clusters= 5, init='k-means++')
kmeans.fit(data_scaled)
prediction = kmeans.predict(data_scaled)

frame = pd.DataFrame(data_scaled)
frame['cluster'] = prediction
frame['cluster'].value_counts()
```
