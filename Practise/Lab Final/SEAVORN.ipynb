```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 8))

# First subplot in a 2x2 grid
plt.subplot(2, 2, 1)
plt.plot([1, 2, 3], [1, 4, 9])
plt.title("Plot 1")

# Second subplot in a 2x2 grid
plt.subplot(2, 2, 2)
plt.plot([1, 2, 3], [1, 2, 3])
plt.title("Plot 2")

# Third subplot in a 2x2 grid
plt.subplot(2, 2, 3)
plt.plot([1, 2, 3], [1, 3, 1])
plt.title("Plot 3")

# Fourth subplot in a 2x2 grid
plt.subplot(2, 2, 4)
plt.plot([1, 2, 3], [2, 2, 2])
plt.title("Plot 4")

plt.tight_layout()  # Adjusts spacing between subplots
plt.show()
```

```python
fig, axes = plt.subplots(2, 2, figsize=(10, 8))

# First plot
axes[0, 0].plot([1, 2, 3], [1, 4, 9])
axes[0, 0].set_title("Plot 1")

# Second plot
axes[0, 1].plot([1, 2, 3], [1, 2, 3])
axes[0, 1].set_title("Plot 2")

# Third plot
axes[1, 0].plot([1, 2, 3], [1, 3, 1])
axes[1, 0].set_title("Plot 3")

# Fourth plot
axes[1, 1].plot([1, 2, 3], [2, 2, 2])
axes[1, 1].set_title("Plot 4")

plt.tight_layout()
plt.show()
```

```python

def calculate_bill(units):
    if(units > 0 and units <= 100):
        return units * 5

    elif(units > 100 and units <= 200):
        units -= 100
        temp = 100 * 5

        return units * 7 + temp

    elif(units > 200 and units <= 300):
        units -= 100
        temp1 = 100 * 5

        units -= 100
        temp2 = 200 * 7

        return units * 9 + temp1 + temp2

    elif(units > 600):
        return units * 12

def imply_tax(bill):
        return bill + (bill * 0.1)

#main
units = int(input("Enter amount of units consumed: "))

bill = calculate_bill(units)

if(bill > 5000):
    tax = imply_tax(bill)
    print("Tax implied: ", tax)

elif(bill <= 5000):
    print("NO TAX IMPLIED !")


print("Electricity Bill: ", bill)
```
