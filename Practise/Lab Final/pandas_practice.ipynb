```python
df = pd.read_csv('tip.csv')
df.head()
```

```python
def last_two(num):
    return str(num)[-2:]

last_two(16.99)
```

```python
df['last_two'] = df['total_bill'].apply(last_two)
df.head()
```

```python
df['total_bill'].mean()
```

```python
def yelp(price):
    if price < 10:
        return '$'
    elif price >= 10 and price < 30:
        return '$$'
    else:
        return '$$$'

df['Expensive'] = df['total_bill'].apply(yelp)
df.head(15)
```

```python
def quality(total_bill,tip):
    if tip/total_bill > 0.25:
        return 'generous'
    else:
        return 'other'
    
df['Tip Quality'] = np.vectorize(quality)(df['total_bill'], df['tip'])
df.head(25)
```
