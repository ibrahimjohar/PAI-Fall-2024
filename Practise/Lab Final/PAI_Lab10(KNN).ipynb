#Supervised Learning


Supervised learning is a type of machine learning where a model is trained using a labeled dataset. It involves learning a mapping from input data (features) to the correct output (labels) by minimizing the difference between the predicted outputs and the actual labels.

#Precision

**Definition:** Precision measures the accuracy of positive predictions. It tells you how many of the items identified as positive are actually correct.

**Formula:**

image.png

```python

precision = precision_score(y_test, y_pred, average='macro')
print(f"Precision: {precision:.2f}")
```

#Recall

**Definition:** Recall measures how well the model identifies all relevant instances. It tells you how many actual positive cases were correctly identified.

**Formula:**
image.png

```python
recall = recall_score(y_test, y_test, average='macro')
print(f"Recall: {recall:.2f}")
```

#F1-SCore

**Definition:** The F1 score is the harmonic mean of precision and recall. It provides a balance between precision and recall, especially useful when the dataset is imbalanced.

**Formula**

image.png

```python
f1 = f1_score(y_test, y_test,average='macro')
print(f"F1 Score: {f1:.2f}")
```
