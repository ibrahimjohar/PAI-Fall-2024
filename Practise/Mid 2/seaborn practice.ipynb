```python
#scatterplots 
#scatter plots show the relationship between two continuous features

import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns

df = pd.read_csv('dm_office_sales.csv')

df.head()
```

```python
sns.scatterplot(x='salary', y='sales', data=df)
```

```python
plt.figure(figsize=(12,4))
sns.scatterplot(x="salary", y="sales", data=df)
```

```python
sns.scatterplot(x='salary', y='sales', data=df, hue='level of education')
```

```python
sns.scatterplot(x="salary", y="sales", data=df, size='salary')
```

```python
sns.scatterplot(x="salary", y='sales', data=df, s=200)
```

```python
sns.scatterplot(x='salary', y='sales', data=df, alpha=0.5)
```

```python
sns.scatterplot(x='salary', y='sales', data=df, style='level of education', hue='level of education')
```

```python
#rugplot 
sns.rugplot(x='salary', data=df)
```

```python
sns.displot(df['salary'])
```

```python
sns.set(style='whitegrid')
sns.displot(df['salary'], bins=20, color='red', edgecolor='black', kde=True)
```

```python
sns.countplot(data=df, x='division')
```

```python
plt.figure(figsize=(9,4))
sns.countplot(data=df, x='division')
```

```python
plt.figure(figsize=(10,4))
sns.countplot(data=df, x='division')
plt.ylim(90,260) #setting boundaries on y axis
```

```python
plt.figure(figsize=(10,4))
sns.countplot(data=df, x='level of education', hue='division')
```

```python
plt.figure(figsize=(10,4))
sns.barplot(data=df, x='level of education', y='salary', estimator=np.mean, errorbar='sd')
```

```python
plt.figure(figsize=(10,4))
sns.barplot(data=df, x='level of education', y='salary', estimator=np.mean, errorbar='sd', hue='division')
```

```python
df = pd.read_csv('StudentsPerformance.csv')
df.head()
```

```python
sns.boxplot(data=df, y='math score')
```

```python
sns.boxplot(data=df, y='math score', x='test preparation course')
```

```python
plt.figure(figsize=(11,6))
sns.boxplot(data=df, y='reading score', x='parental level of education')
```

```python
plt.figure(figsize=(11,6))
sns.boxplot(data=df, y='reading score', x='parental level of education', hue='test preparation course')
```

```python
plt.figure(figsize=(8,4), dpi=200)
sns.swarmplot(data=df, x='math score', y='gender', size=3)
```

```python
plt.figure(figsize=(8,4),dpi=200)
sns.swarmplot(data=df, x='math score', y='gender', hue='test preparation course', size=2)
plt.legend(bbox_to_anchor=(1.35,0.5))
```

```python
plt.figure(figsize=(8,4), dpi=200)
sns.swarmplot(data=df, x='math score', y='gender', hue='test preparation course', size=2, dodge=True)
plt.legend(bbox_to_anchor=(1.35,0.5))
```

```python
sns.boxenplot(data=df, x='math score', y='test preparation course')
```

```python
sns.jointplot(data=df, x='math score', y='reading score')
```

```python
sns.jointplot(data=df, x='math score', y='reading score', alpha=0.5)
```

```python
sns.jointplot(data=df, x='math score', y='reading score', kind='kde')
```

```python
sns.jointplot(data=df, x='math score', y='reading score', kind='kde', fill=True)
```

```python
sns.jointplot(data=df, x='math score', y='reading score', hue='gender')
```

```python
sns.pairplot(df)
```

```python
#kind options are "point", "bar", " strip", "swarm", "box","violen", or "boxen"
sns.catplot(x='gender', y='math score', data=df, kind='box')
```

```python
sns.catplot(x='gender',y='math score', data=df, kind='box', row='lunch')
```

```python
sns.catplot(x='gender',y='math score', data=df, kind='box', row='lunch', col='test preparation course')
```
