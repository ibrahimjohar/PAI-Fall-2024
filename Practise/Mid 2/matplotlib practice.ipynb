```python
#subplots

x = [1,2,3,4]
y = [1,2,3,4]

#create a 2x2 grid of subplots and plot on the first subplot
plt.subplot(2,2,1)
plt.plot(x,y,color="r") #plot with red line

#plot on the second subplot in the 2x2 grid
plt.subplot(2,2,2)
plt.plot(x,y,color="r")

#plot on the third subplot in the 2x2 grid
plt.subplot(2,2,3)
plt.plot(x,y,color="r")

#plot on the fourth subplot in the 2x2 grid
plt.subplot(2,2,4)
plt.plot(x,y,color="r")

plt.show()
```
