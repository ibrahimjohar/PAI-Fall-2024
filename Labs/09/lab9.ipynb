```python
#q01
import matplotlib.pyplot as plt

students = ['John', 'Sophia', 'Liam', 'Olivia', 'Noah', 'Emma', 'James', 'Ava', 'Lucas', 'Isabella',
            'Mason', 'Mia', 'Ethan', 'Charlotte', 'Logan', 'Amelia', 'Aiden', 'Harper', 'Jacob', 'Evelyn']

heights = [160, 165, 170, 155, 172, 168, 177, 162, 180, 175, 178, 164, 169, 173, 165, 170, 176, 168, 158, 167]


plt.figure(figsize=(10, 5)) 
plt.bar(students, heights, color=['blue', 'green', 'red', 'orange', 'purple', 'cyan', 'pink', 'yellow', 
                                  'lightgreen', 'lightblue', 'gray', 'magenta', 'lightcoral', 'lime', 
                                  'violet', 'gold', 'turquoise', 'chocolate', 'navy', 'salmon'])

plt.title("Student Heights (in cm) - Bar Chart")
plt.xlabel("Students")
plt.ylabel("Heights (cm)")
plt.xticks(rotation=45, ha='right')  
plt.tight_layout()  


plt.show()


plt.figure(figsize=(8, 8))  
plt.pie(heights, labels=students, autopct='%1.1f%%', startangle=140, colors=['blue', 'green', 'red', 
         'orange', 'purple', 'cyan', 'pink', 'yellow', 'lightgreen', 'lightblue', 'gray', 'magenta', 
         'lightcoral', 'lime', 'violet', 'gold', 'turquoise', 'chocolate', 'navy', 'salmon'])

plt.title("Student Heights Distribution - Pie Chart")

plt.show()
```

```python
#q02

import matplotlib.pyplot as plt

cities = ['Tokyo', 'Delhi', 'Shanghai', 'Sao Paulo', 'Mumbai', 'Cairo', 'Beijing', 'Dhaka', 'Mexico City', 'Osaka']
populations = [37400068, 28514000, 25582000, 21650000, 20411000, 20076000, 20035455, 19578000, 21581000, 19281000]


plt.figure(figsize=(10, 6))  
plt.barh(cities, populations, color='skyblue')

plt.title('City Populations')
plt.xlabel('Population')
plt.ylabel('Cities')


plt.grid(axis='x', linestyle='--', alpha=0.6)


for index, value in enumerate(populations):
    plt.text(value, index, f'{value:,}', va='center', ha='left')


plt.tight_layout()  
plt.show()
```

```python
#q04
import matplotlib.pyplot as plt

ages = [22, 24, 19, 27, 29, 32, 21, 25, 34, 23, 28, 31, 35, 40, 37, 26, 33, 30, 22, 29]

age_groups = {
    '18-25': 0,
    '26-30': 0,
    '31-35': 0,
    '36 and above': 0
}

for age in ages:
    if 18 <= age <= 25:
        age_groups['18-25'] += 1
    elif 26 <= age <= 30:
        age_groups['26-30'] += 1
    elif 31 <= age <= 35:
        age_groups['31-35'] += 1
    elif age >= 36:
        age_groups['36 and above'] += 1

labels = list(age_groups.keys()) 
sizes = list(age_groups.values())  

plt.figure(figsize=(8, 8))
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90, colors=plt.get_cmap('Pastel1').colors)

plt.title('Distribution of Student Ages by Age Group')
plt.axis('equal')  

plt.show()
```

```python
#q05
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv('students.csv')
gender_counts = data['Gender'].value_counts()

plt.figure(figsize=(6, 6))
plt.pie(gender_counts, labels=gender_counts.index, autopct='%1.1f%%', startangle=140, colors=['skyblue', 'lightpink'])
plt.title('gender distribution of students')
plt.show()
```

```python
#q06
import numpy as np
import matplotlib.pyplot as plt

mathsMarks = np.random.randint(0, 100, 10)
scienceMarks = np.random.randint(0, 100, 10)

plt.scatter(range(10), mathsMarks, color='blue', label='Maths')
plt.scatter(range(10), scienceMarks, color='green', label='Science')

plt.xlabel('student index')
plt.ylabel('marks')
plt.legend()
plt.title('comparison of maths and science marks')
plt.show()
```

```python
#q07
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('Global_sea_level_rise.csv')

if len(df) > 100:
    df = df.iloc[-100:]

plt.figure(figsize=(10, 6))
plt.scatter(df['year'], df['mmfrom1993-2008average'], color='blue', label='Average Sea Level Rise')

plt.xlabel('year')
plt.ylabel('sea level rise (mm)')
plt.title('sea level rise in the past 100 years')
plt.legend()
plt.grid(True)

plt.show()
```
