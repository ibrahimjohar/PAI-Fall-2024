```python
#q01
import matplotlib.pyplot as plt

students = ['John', 'Sophia', 'Liam', 'Olivia', 'Noah', 'Emma', 'James', 'Ava', 'Lucas', 'Isabella',
            'Mason', 'Mia', 'Ethan', 'Charlotte', 'Logan', 'Amelia', 'Aiden', 'Harper', 'Jacob', 'Evelyn']

heights = [160, 165, 170, 155, 172, 168, 177, 162, 180, 175, 178, 164, 169, 173, 165, 170, 176, 168, 158, 167]


plt.figure(figsize=(10, 5)) 
plt.bar(students, heights, color=['blue', 'green', 'red', 'orange', 'purple', 'cyan', 'pink', 'yellow', 
                                  'lightgreen', 'lightblue', 'gray', 'magenta', 'lightcoral', 'lime', 
                                  'violet', 'gold', 'turquoise', 'chocolate', 'navy', 'salmon'])

plt.title("Student Heights (in cm) - Bar Chart")
plt.xlabel("Students")
plt.ylabel("Heights (cm)")
plt.xticks(rotation=45, ha='right')  
plt.tight_layout()  


plt.show()


plt.figure(figsize=(8, 8))  
plt.pie(heights, labels=students, autopct='%1.1f%%', startangle=140, colors=['blue', 'green', 'red', 
         'orange', 'purple', 'cyan', 'pink', 'yellow', 'lightgreen', 'lightblue', 'gray', 'magenta', 
         'lightcoral', 'lime', 'violet', 'gold', 'turquoise', 'chocolate', 'navy', 'salmon'])

plt.title("Student Heights Distribution - Pie Chart")

plt.show()
```

```python
#q02

import matplotlib.pyplot as plt

cities = ['Tokyo', 'Delhi', 'Shanghai', 'Sao Paulo', 'Mumbai', 'Cairo', 'Beijing', 'Dhaka', 'Mexico City', 'Osaka']
populations = [37400068, 28514000, 25582000, 21650000, 20411000, 20076000, 20035455, 19578000, 21581000, 19281000]


plt.figure(figsize=(10, 6))  
plt.barh(cities, populations, color='skyblue')

plt.title('City Populations')
plt.xlabel('Population')
plt.ylabel('Cities')


plt.grid(axis='x', linestyle='--', alpha=0.6)


for index, value in enumerate(populations):
    plt.text(value, index, f'{value:,}', va='center', ha='left')


plt.tight_layout()  
plt.show()
```

```python
import matplotlib.pyplot as plt


flavors = ['Vanilla', 'Chocolate', 'Strawberry', 'Mint', 'Cookie Dough', 'Mango', 'Pistachio', 'Lemon']
scoops_sold = [150, 100, 90, 60, 80, 70, 50, 40]


plt.figure(figsize=(8, 8))  
plt.pie(scoops_sold, labels=flavors, autopct='%1.1f%%', startangle=90, colors=plt.cm.Paired.colors)

plt.title('Ice Cream Sales Distribution')


plt.show()
```
